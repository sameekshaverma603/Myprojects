```python
#Will train our model using linear regression as its corelated with only one variable 
from sklearn.model_selection import train_test_split
```

```python
x=data['TV']
y=data['Sales']
```

```python
x.shape
```

```python
y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
model=LinearRegression()
model.fit(pd.DataFrame(x_train),pd.DataFrame(y_train))
```

```python
y_pred=model.predict(pd.DataFrame(x_test))
print(y_pred)
```

```python
print(model.coef_)
```

```python
print(model.intercept_)
```

```python
plt.plot(y_pred,color='blue')
plt.show()
```

```python
#will fit a straight line to see how well our model is performing 
plt.scatter(x_test , y_test)
plt.plot(x_test,7.20655455 + 0.05483488 * x_test,'r')
plt.show()
```
