```python
#check for null values 
data.isnull().sum()
```

```python
#will use standard scaling for feature scaling the amount column to have same range values as other columns;
from sklearn.preprocessing import StandardScaler  
```

```python
sc=StandardScaler()
data['Amount']=sc.fit_transform(pd.DataFrame(data['Amount']))
```

```python
data.head()
```

```python
#Drop Time column 
data=data.drop(['Time'],axis=1)
```

```python
data.head()
```

```python
data.duplicated().any()
```

```python
data=data.drop_duplicates()
```

```python
data.shape
```

```python
data['Class'].value_counts() 
```

```python
#Clearly we can see that its imbalanced dataset 
import seaborn as sns 
```

```python
sns.countplot(data['Class'])
```

```python
#Handling the imbalanced dataset using Undersampling 
normal=data[data['Class']==0]
fraud=data[data['Class']==1]
```

```python
normal.shape
```

```python
fraud.shape
```

```python
normal_sample=normal.sample(n=473)
```

```python
normal_sample.shape
```

```python
new_data=pd.concat([normal_sample,fraud],ignore_index=True)
```

```python
new_data['Class'].value_counts()
```

```python
new_data.head()
```

```python
x=new_data.drop('Class',axis=1)
y=new_data['Class']
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
```

```python
log=LogisticRegression()
log.fit(x_train,y_train)
```

```python
y_pred1=log.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred1)
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
precision_score(y_test,y_pred1)
```

```python
recall_score(y_test,y_pred1)
```

```python
f1_score(y_test,y_pred1)
```

```python
#Decision Tree 
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
y_pred2=dt.predict(x_test)
```

```python
accuracy_score(y_test,y_pred2)
```

```python
precision_score(y_test,y_pred2)
```

```python
recall_score(y_test,y_pred2)
```

```python
f1_score(y_test,y_pred1)
```

```python
#Random forest
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_pred3=rf.predict(x_test)
```

```python
accuracy_score(y_pred3,y_test)
```

```python
precision_score(y_pred3,y_test)
```

```python
recall_score(y_pred3,y_test)
```

```python
f1_score(y_pred3,y_test)
```

```python
#Lets create a DataFrame for Accuracies 
final_data=pd.DataFrame({'Models':['LR','DT','RF'],
            'Accuracy':[accuracy_score(y_pred1,y_test)*100,
                         accuracy_score(y_pred2,y_test)*100,
                         accuracy_score(y_pred3,y_test)*100
                        ]})
final_data
```

```python
sns.barplot(final_data['Models'],final_data['Accuracy'])
```
